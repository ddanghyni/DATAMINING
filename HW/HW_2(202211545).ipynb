### 이상치 제거

```python
def remove_outlier_by_iqr(df, col, k=1.5):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - k * iqr
    upper_bound = q3 + k * iqr
    return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]



df_clean = df.copy()

for col in df_clean.columns:
    df_clean = remove_outlier_by_iqr(df_clean, col, k=1.5)

print(df.shape, "→", df_clean.shape)
```

```python
import pandas as pd
from functools import reduce

def classwise_iqr_filter(
    df: pd.DataFrame,
    target_col: str = "Outcome",
    k: float = 1.5,
    numeric_only: bool = True,
):
    """
    클래스별 IQR 기반 이상치 제거.
    """
    kept_indices = []   # 각 클래스에서 살아남은 인덱스 보관

    for cls, grp in df.groupby(target_col):
        # 1) 숫자형 열(타깃 제외) 선택
        if numeric_only:
            num_cols = grp.select_dtypes(include="number").columns.drop(target_col)
        else:
            num_cols = grp.columns.drop(target_col)

        # 2) IQR 경계 계산
        Q1 = grp[num_cols].quantile(0.25)
        Q3 = grp[num_cols].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - k * IQR
        upper = Q3 + k * IQR

        # 3) 하나라도 경계 밖이면 이상치로 간주 → 제외
        mask = ~((grp[num_cols] < lower) | (grp[num_cols] > upper)).any(axis=1)
        kept_indices.append(grp[mask].index)

    # 4) 클래스별로 남은 인덱스를 모두 합치기
    all_kept_idx = reduce(pd.Index.union, kept_indices)

    return df.loc[all_kept_idx].copy()

# ────────────────────────────────
# 사용 예시
df_class_clean = classwise_iqr_filter(df, target_col="Outcome", k=1.5)
print(df.shape, "→", df_class_clean.shape)
```

```python
"""
Reusable function for logistic regression with RFE feature selection on tabular data.

Encapsulates:
1. Train/test split
2. Standardization
3. Recursive Feature Elimination (RFE)
4. Logistic regression model fitting
5. Evaluation (confusion matrix & classification report)

Example usage:
--------------
from logistic_rfe_pipeline import run_logistic_rfe

run_logistic_rfe(df, target_col='Outcome', n_features=10, test_size=0.2)
"""

from typing import Optional

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

def run_logistic_rfe(
    df: pd.DataFrame,
    target_col: str,
    n_features: int = 10,
    test_size: float = 0.2,
    random_state: int = 42,
    solver: str = "liblinear",
    max_iter: int = 10000,
    print_report: bool = True
):
    """Runs logistic regression + RFE pipeline on df.

    Parameters
    ----------
    df : pd.DataFrame
        Input data containing features and target.
    target_col : str
        Name of the target column.
    n_features : int, default 10
        Number of features to select in RFE.
    test_size : float, default 0.2
        Test set proportion.
    random_state : int, default 42
        Random seed for reproducibility.
    solver : str, default 'liblinear'
        Solver to use for LogisticRegression.
    max_iter : int, default 10000
        Maximum iterations for logistic solver.
    print_report : bool, default True
        Whether to print confusion matrix and classification report.
    """
    # 1. Split
    X = df.drop(columns=target_col)
    y = df[target_col]

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, stratify=y
    )

    # 2. Scale
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # 3. RFE
    base_estimator = LogisticRegression(
        random_state=random_state, solver=solver, max_iter=max_iter
    )
    selector = RFE(base_estimator, n_features_to_select=n_features)
    selector = selector.fit(X_train_scaled, y_train)

    print("선택된 변수 (True는 선택된 변수):\n", selector.support_)
    print("변수의 순위 (1-선택된 변수, 숫자가 클수록 덜 중요한 변수):\n", selector.ranking_)

    # 4. Select features
    X_train_sel = selector.transform(X_train_scaled)
    X_test_sel = selector.transform(X_test_scaled)

    # 5. Train final model
    model = LogisticRegression(random_state=random_state, max_iter=max_iter)
    model.fit(X_train_sel, y_train)
    y_pred = model.predict(X_test_sel)

    # 6. Evaluation
    if print_report:
        print("\nConfusion Matrix:")
        print(confusion_matrix(y_test, y_pred))
        print("\nClassification Report:")
        print(classification_report(y_test, y_pred))

    return model, selector
```

```python
run_logistic_rfe(df,    target_col='Outcome',
    n_features=10,
    test_size=0.2,
    random_state=42,
    solver='liblinear',
    max_iter=10000,
    print_report=True
)
```
