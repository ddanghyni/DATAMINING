# 트리 기반 분류 Task에서 분할 기준
Gini Index vs. Cross Entropy  
이진 분할의 예

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('retina')
```

```python
plt.rcParams['font.family'] = 'AppleGothic'
#plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
p = np.linspace(0.001, 0.999, 1000)
```

```python
gini = 2 * p * (1 - p) # p(1-p) + p(1-p)
```

```python
entropy = - (p * np.log2(p) + (1 - p) * np.log2(1 - p))
```

```python
plt.figure(figsize=(8, 5))
plt.plot(p, gini, label='Gini Index', linewidth=2)
plt.plot(p, entropy, label='Entropy', linewidth=2)
plt.xlabel('클래스 1의 확률 (p)')
plt.ylabel('지표 값')
plt.title('지니 지수 vs. 교차 엔트로피')
plt.legend()
plt.grid(True)
plt.show()
```
